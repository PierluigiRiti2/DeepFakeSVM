```python
import cv2
import numpy as np
# import matpotlib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
def draw_histogram(image_file):
    #image read function
    img=mpimg.imread(image_file)
    #image sclicing into 2D. 
    x=img[:,:,0]
    # x co-ordinate denotation. 
    plt.xlabel("Value")
    # y co-ordinate denotation.
    plt.ylabel("pixels Frequency")
    # title of an image .
    plt.title("Original Image")
    # imshow function with comperision of gray level value.
    plt.imshow(x,cmap="gray")
    #plot the image on a plane.
    plt.show()

    plt.title("HIstogramm for given Image'  ")
    plt.xlabel("Value")
    plt.ylabel("pixels Frequency")
    #hist function is used to plot the histogram of an image.
    plt.hist(x)
```

```python
#/home/pierluigiriti/deepfakes-detection-master/image_save/original/
def read_image(file_read, path_save):
    # Create the VideCapture to capture the video in the path
    # read the original file
    original_video = cv2.VideoCapture(file_read) #'/home/pierluigiriti/deepfakes-detection-master/VidTIMIT/VidTIMIT/fadg0/sa1.avi')

    # check if the camera is correctly open
    if(original_video.isOpened()==False):
        print("Error open video or streaming file")

    i = 1

    while(original_video.isOpened()):
        # read the video frame by frame
        ret, frame = original_video.read()

        if ret == True:
            file_name = f"{path_save}/original_{i}.jpg"
            # save the frame for the image
            cv2.imwrite(file_name, 
                        frame)
            
            draw_histogram(file_name)

        else:
            break

        # create the counter
        i += 1

    original_video.release()

    cv2.destroyAllWindows()
```

```python
# Read the Original file and save in the path
read_image('/home/pierluigiriti/deepfakes-detection-master/VidTIMIT/VidTIMIT/fadg0/sa1.avi',
          '/home/pierluigiriti/deepfakes-detection-master/image_save/original/')
```
